2.1 Calculate the total revenue for the store (sum of all quantities ×
price per item).

```python
total_revenue = sum(order["quantity"] * order["price_per_item"] for order in sales_data)
print(f"\nTotal revenue: {total_revenue:,.2f}")
```

2.2 Identify all orders where the total order value (quantity × price
per item) exceeds ₹2,000. Store the result as a new list.
2.3 Find the item with the highest total quantity sold across all
orders.

```python
high_value_orders = [order for order in sales_data if (order['quantity'] * order['price_per_item']) > 2000]
for order in high_value_orders:
    print(order)

print(f"Count of high value orders: {len(high_value_orders)}")
```

2.3 Find the item with the highest total quantity sold across all
orders.

```python
quantity_sold = {}
for order in sales_data:
    item = order['item']
    quantity_sold[item] = quantity_sold.get(item, 0) + order['quantity']
print(quantity_sold)
```

3.1 Write a function apply_
discount(orders, discount
_
rate) that takes
a list of orders and applies a percentage discount (e.g., 10% = 0.10) to
the price
_per
_
item for each order.
Return a new list of orders (do not modify the original list).

```python
def apply_discount(orders, discount_rate):
  discounted_orders = [order.copy() for order in sales_data]
  for order in discounted_orders:
      order['price_per_item'] *= (1 - discount_rate)

  return discounted_orders
```

4.2 Create a dictionary grouping all orders by customer using a
dictionary comprehension.

```python
particular_customers = {order['customer'] for order in sales_data}


customer_order = {
    cust: [order for order in sales_data if order['customer'] == cust]
    for cust in particular_customers
}

print("Manu's Orders count:", len(customer_order['Manu']))
print("Gopika's Orders count:", len(customer_order['Gopika']))
```
